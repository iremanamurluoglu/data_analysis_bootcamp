```python
import numpy as np
import pandas as pd
import seaborn as sns
mall=pd.read_csv("/content/Mall_Customers.csv")
mall.head()
```

```python
mall.tail()
```

```python
mall.describe()
```

```python
import matplotlib.pyplot as plt
plt.scatter(mall["Spending Score (1-100)"],mall["Annual Income (k$)"], c=mall.Age)
plt.xlabel("Spending Score (1-100)")
plt.ylabel("Annual Income (k$)")
plt.colorbar()
```

```python
mall["Greediness"] = mall["Spending Score (1-100)"]/mall["Annual Income (k$)"]
```

```python
mall["Greediness"].hist(bins=50)
```

```python
min_age = mall["Age"].min()
max_age = mall["Age"].max()
```

```python
mall["Age"].hist(bins= max_age - min_age)
```

```python
mall["Gender"].hist(bins= 2)
```

```python
sns.boxplot(data=mall,x='Gender',y='Spending Score (1-100)')
```

```python
plt.figure(figsize=(15,8))
plt.scatter(mall['Annual Income (k$)'], mall['Spending Score (1-100)'],c=mall.Age,alpha=0.6)
plt.title('Distribution of Age based on Annual Income and Spending Score', fontsize = 15)
plt.xlabel('Annual Income', fontsize = 12)
plt.ylabel('Spending Score', fontsize = 12)
plt.colorbar()
plt.show()
```
