```python
import numpy as np
import pandas as pd
import seaborn as sns
titanic=sns.load_dataset("titanic")
titanic.head()
```

```python
#1- pivot table oluştur. survived, class ve embarked columnlarını kullanarak. 
#Tüm embarked ve classlarda kaç kişinin survive edip etmediğini bul,Ayrıca ortalamasını bul.
titanic.pivot_table("survived",index="class",columns="embarked",aggfunc=["mean","count"])

#Hayatta kalma ortalaması olarak first classta olan ve Cherbourg limanından binenler daha yüksektir (diğer limanlara göre)
#Southampton limanından binen insanlarda en fazla yolcu third class 'ta mevcuttur.
#Queenstown limanından binen yolcularda en fazla hayatta kalma oranı second class yolcularındadır.
#First classta en çok hayatta kalan insan sayısı Southampton limanından binenlerdendir.
```

```python
#2-Embark town , 0-15,15-24,24-36,36-45,45-60,60 + yaş aralıkları ve cinsiyet e göre grupla survive edip/etmeyen kişi sayısını ve ortalamasını bul.
age=pd.cut(titanic["age"],[0,15,24,36,45,60,titanic["age"].max()])
age
```

```python
titanic.pivot_table("survived", ["sex",age],"embark_town",aggfunc=["mean","count"])

#Female(kadın) olarak 60 ve 80 yaş aralığında sadece Southampton limanından binen vardır.
#0-15 yaş aralığında Cherbourg limanından binenlerde kadınların erkeklere göre hayatta kalma oranı daha yüksektir.
#Kadınlarda 24-36 yaş aralığında Southampton limanından binenlerin sayısı diğer limanlara ve diğer yaş aralıklarına göre daha fazladır.
#Erkekler de 0-15 yaş aralığında Southampton limanından binenlerden hayatta kalan insan sayısı diğer limanlardan daha yüksektir.
```

```python
#3-  Groupby kullanarak embark town ve fare arasında bir ilişki var mı yok mu bulamaya çalış.
titanic.groupby("embark_town")["fare"].describe()

#Cherbourg limanından binen insanların ortalama bilet fiyatı daha yüksektir diğer limanlaran binenlere göre.
#Cherbourg limanından en fazla 512.3292 ücret mevcutken en düşük 90 ücret Queenstown limanındadır.
#Limanlara göre ücrette farklılıklar mevcuttur en fazla Cherbourg  sonra Southampton ve en düşük ise Queenstown limanından ücret kesilmiştir, şehir zenginliği ile doğru orantılı ücret fiyatı vardır.
```

```python
#4- Cinsiyete göre hangi limandan (embark town) kaçar kişi binmiş? 
titanic.groupby(["sex","embark_town"]).size()

#Kadınlar ve erkekler de en fazla Southampton limanından binilmiştir fakat Southampton limanından binen erkek sayısı kadın sayısının 2 katından fazladır.
#En az Queenstown limanından insan binmiştir ve binen erkek sayısı kadına göre daha fazladır bu limanda.
```

```python
#Classlara  göre hangi limandan (embark town) kaçar kişi binmiş? 
titanic.groupby(["class","embark_town"]).size()

#First, Second ve Third class'ta en fazla Southampton limanından bilmiştir.
#First ve Second class'ta en az Queenstown limanından binilmiştir fakat Third class ise en az Cherbourg limanından binilmiştir.
#En fazla yolcu Thirde Class' ta mevcuttur.
```

```python
#Ve hangi cinsiyetten ve classtan kaç kişi binmiş? ( second class  5 erkek Cherbourg limanından binmiş gibi)
titanic.groupby(["sex","class","embark_town"]).size()

#Kadınlarda first class ta olanlar en fazla Southampton limanından binmiştir.
#Southampton limanından binen kadınlar en fazla Third class ta mevcuttur.
#Erkeklerde tüm classlar da Queenstown limanından en az sayıda binilmiştir.
```

```python
#5-sibsp yi şu şekilde değiştir. Sıfırlar, birler ve ikiler aynı kalsın  üçten büyük olanların hepsini 3  olarak değiş.
titanic["sibsp"].value_counts()
```

```python
titanic.loc[titanic['sibsp'] > 3,["sibsp"]]=3
titanic["sibsp"].value_counts()
```

```python
#sibsp  survived ilişkisi (survive eden/etmeyen sayısı ve mean) 
survived_alive_or_notalive=titanic.groupby("sibsp")["survived"].aggregate(["count","mean"])
survived_alive_or_notalive["not_alive_mean"]=1-survived_alive_or_notalive["mean"]
survived_alive_or_notalive["alive_count"]=survived_alive_or_notalive["mean"]*survived_alive_or_notalive["count"]
survived_alive_or_notalive["not_alive_count"]=survived_alive_or_notalive["count"]-survived_alive_or_notalive["alive_count"]
survived_alive_or_notalive

#sibsp 0 olan da toplam 608 insan vardır ve 210 insan hayatta kalmıştır. Hayatta kalma oranı yaklaşık olarak 0.34.
#sibsp 1 hayatta kalma oranı en yüksektir.
#sibsp 3 hayatta kalma oranı en düşüktür.
#sibsp 1 de hayatta kalma oranı 0.53 hayatta kalan insan sayısı 112 dir. sibsp 0 da hayatta kalma oranı 0.34 hayatta kalan insan sayısı 210dur.
```

```python
#sibsp ve sex e göre  survive eden/etmeyen sayısı ve mean
survived_alive_or_notalive=titanic.groupby(["sibsp","sex"])["survived"].aggregate(["count","mean"])
survived_alive_or_notalive["not_alive_mean"]=1-survived_alive_or_notalive["mean"]
survived_alive_or_notalive["alive_count"]=survived_alive_or_notalive["mean"]*survived_alive_or_notalive["count"]
survived_alive_or_notalive["not_alive_count"]=survived_alive_or_notalive["count"]-survived_alive_or_notalive["alive_count"]
survived_alive_or_notalive

#sibsp 0 değerinde kadınlarda 0.78 hayatta kalma oranı varken erkeklerde 0.16 dır.
#sibsp 1 değerinde kadınlarda 106 kişi mevcutken hayatta kalan 80 kişidir. Erkelerde 103 kişi mevcutken hayatta kalan 32 kişidir. Bu durumda görüldüğü gibi kadınlar sibsp1 de daha fazla hayatta kalmıştır.
#sibsp 0 değerinde en fazla insan mevcutken sibsp 2 değerinde en az insan mevcuttur. sibsp 0 ve sibsp 1 erkek sayısı daha fazladı(titanic olmadan önce) fakat hayatta kalan kadın sayısı daha fazladır.
```

```python
#sibsp ve age e göre  survive eden/etmeyen sayısı ve mean
age=pd.cut(titanic["age"],[0,15,24,36,45,60,titanic["age"].max()])
survived_alive_or_notalive=titanic.groupby(["sibsp",age])["survived"].aggregate(["count","mean"])
survived_alive_or_notalive["not_alive_mean"]=1-survived_alive_or_notalive["mean"]
survived_alive_or_notalive["alive_count"]=survived_alive_or_notalive["mean"]*survived_alive_or_notalive["count"]
survived_alive_or_notalive["not_alive_count"]=survived_alive_or_notalive["count"]-survived_alive_or_notalive["alive_count"]
survived_alive_or_notalive

#sibsp 0 değerinde en fazla hayatta kalma oranı 0-15 yaş aralığında mevcuttur.
#sibsp 1 değerinde hayatta kalmama oranı en fazla 60-80 yaş aralığında mevcuttur.
#sibsp 2 değerinde en fazla 15-24 yaş aralığında insan vardır ve hayatta kalma oranı 0.36 dır.0-15 yaş aralığında daha az insan varken haytta kalma oranı 0.83.Sonuç olarak 0-15 yaş aralığında,15-24 yaş aralığına göre daha fazla insan hayatta kalmıştır.
#sibsp 3 değerinde 36 yaşın üstünde insan bulunmamaktadır.
```
