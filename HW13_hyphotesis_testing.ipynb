One Proportion Test (Z-test)

```python
#One Proportion Test (Z-test):İkili bir kategorik değişken için iki grup arasındaki oranları karşılaştırmak.
#kalp hastalığı olan erkeklerin oranı, varsayılan bir orandan önemli ölçüde farklıysa.
from statsmodels.stats.proportion import proportions_ztest

# One Proportion Testi için ilgili sütun(erkekler/kadınlar için 'cinsiyet' ve kalp hastalığı için 'hedef')
male_heart_disease = heart[(heart['sex'] == 1) & (heart['target'] == 1)]
n_males = len(heart[heart['sex'] == 1])
p_hypothesis = 0.5  # Varsayılan oran (eşit oranlar için %50)

# Perform the one proportion Z-test
z_stat, p_value = proportions_ztest(len(male_heart_disease), n_males, p_hypothesis)

print("Z-statistic:", z_stat)
print("P-value:", p_value)
#negatif Z-statistiği, birinci gruptaki özelliğin daha küçük olduğunu ve düşük P-değerinin, iki grup arasındaki özelliğin anlamlı bir fark gösterdiğini gösterdiğini söyleyebiliriz. Yani, iki grup arasındaki özelliğin ortalama değerlerinde gerçekten anlamlı bir fark olduğu sonucuna varabiliriz
```

Two Proportions Test (Z-test)

```python
#İkili bir kategorik değişken için iki bağımsız grup arasındaki oranları karşılaştırmak.
#Kalp hastalığı olan kadın ve erkek oranlarında ise anlamlı bir fark vardır.
female_heart_disease = heart[(heart['sex'] == 0) & (heart['target'] == 1)]
n_females = len(heart[heart['sex'] == 0])

# Perform the two proportions Z-test
count = np.array([len(male_heart_disease), len(female_heart_disease)])
nobs = np.array([n_males, n_females])

z_stat, p_value = proportions_ztest(count, nobs)

print("Z-statistic:", z_stat)
print("P-value:", p_value)
```

One Sample t-test

```python
#Bir örneğin ortalamasını bilinen bir popülasyon ortalamasına veya varsayılan bir değere karşı karşılaştırmak.
#ortalama kolesterol seviyesi, varsayılan bir popülasyon ortalamasından önemli ölçüde farklıysa.
from scipy.stats import ttest_1samp

# Select the relevant column for the one-sample t-test (e.g., 'chol' for cholesterol levels)
sample_mean = heart['chol'].mean()
pop_mean_hypothesis = 240  # Hypothesized population mean

# Perform the one-sample t-test
t_stat, p_value = ttest_1samp(heart['chol'], pop_mean_hypothesis)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

Two Sample t-test

```python
#İki bağımsız grubun ortalamalarını karşılaştırmak.
#kalp hastalığı olan ve olmayan hastalar arasında kolesterol düzeylerinde anlamlı bir fark varsa.
# Select the relevant columns for the two-sample t-test (e.g., 'chol' for cholesterol levels and 'target' for heart disease)
from scipy.stats import ttest_ind
chol_with_heart_disease = heart[heart['target'] == 1]['chol']
chol_without_heart_disease = heart[heart['target'] == 0]['chol']

# Perform the two-sample t-test
t_stat, p_value = ttest_ind(chol_with_heart_disease, chol_without_heart_disease)

print("T-statistic:", t_stat)
print("P-value:", p_value)

```

One Proportion Test (Z-test)

```python
#İkili bir kategorik değişken için iki grup arasındaki oranları karşılaştırmak.
#bir konuda 'evet' oyu veren 'demokratların' oranı, varsayılan bir orandan önemli ölçüde farklıysa.
from statsmodels.stats.proportion import proportions_ztest
# Choose a relevant column for the one proportion test (e.g., 'education', 'budget', 'immigration', etc.)
chosen_column = ' education-spending'

# Select the relevant data for Democrats and 'yes' votes
democrats_yes = votes[(votes['Class Name'] == 'democrat') & (votes[chosen_column] == 'y')]
n_democrats = len(votes[votes['Class Name'] == 'democrat'])
p_hypothesis = 0.5  # Hypothesized proportion (e.g., 50% for equal proportions)

# Perform the one proportion Z-test
z_stat, p_value = proportions_ztest(len(democrats_yes), n_democrats, p_hypothesis)

print("Z-statistic:", z_stat)
print("P-value:", p_value)
```

Two Proportions Test (Z-test)

```python
#İkili bir kategorik değişken için iki bağımsız grup arasındaki oranları karşılaştırmak.
#bir konuda 'evet' oyu veren 'demokratlar' ve 'cumhuriyetçiler' oranlarında anlamlı bir fark varsa.
# Select the relevant data for Republicans and 'yes' votes
republicans_yes = votes[(votes['Class Name'] == 'republican') & (votes[chosen_column] == 'y')]
n_republicans = len(votes[votes['Class Name'] == 'republican'])

# Perform the two proportions Z-test
count = np.array([len(democrats_yes), len(republicans_yes)])
nobs = np.array([n_democrats, n_republicans])

z_stat, p_value = proportions_ztest(count, nobs)

print("Z-statistic:", z_stat)
print("P-value:", p_value)
```

One Sample t-test

```python
#Bir örneğin ortalamasını bilinen bir popülasyon ortalamasına veya varsayılan bir değere karşı karşılaştırmak.
#belirli bir konu için ortalama oylama sıklığı varsayımsal bir değerden önemli ölçüde farklıysa.
from scipy.stats import ttest_1samp

# Choose a relevant column for the one-sample t-test (e.g., 'education', 'budget', 'immigration', etc.)
chosen_column = ' education-spending'

# Select the relevant data for the one-sample t-test
sample_mean = votes[chosen_column].map({'y': 1, 'n': -1, '?':0}).mean()
pop_mean_hypothesis = 0.5  # Hypothesized population mean (e.g., 0.5 for equal proportions)

# Perform the one-sample t-test
t_stat, p_value = ttest_1samp(votes[chosen_column].map({'y': 1, 'n': -1, '?':0}), pop_mean_hypothesis)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```
